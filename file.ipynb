```python
import pandas as pd
df2021=pd.read_csv('Emission2021.csv')
print(df2021['ImoNumber'].nunique())
df2020=pd.read_csv('Emission2020.csv')
print(df2020['ImoNumber'].nunique())
df2019=pd.read_csv('Emission2019.csv')
print(df2019['ImoNumber'].nunique())
df_concat=pd.concat([df2021,df2020,df2019])
df_concat.to_csv('Emission_concats.csv',index=False)
print(df_concat['ImoNumber'].nunique())
```
